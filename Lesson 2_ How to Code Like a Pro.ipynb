# Errors

## What do errors look like?

Everyone fucks up. Especially when they're writing Python code. For that reason, Python has a neat system to try and give you as much information about what you've done wrong.

````python
daniel is cool
````

The above code snippet didn't work for a few reasons. Python tried running this code, but when it encountered the term `daniel`, it looked at all the defined variables and found nothing. So because it doesn't know what `daniel` is equal to (i.e. we haven't defined it yet), Python threw a `NameError`.

## Whenever you see a **NameError**, that means you tried using a variable that has not been defined yet.

````python
daniel = 3
daniel is cool
````

If we go ahead and define `daniel`, you can notice that we still get the same NameError, but this time for the variable `cool`.

````python
daniel = 10
cool = 10
daniel is cool # should be True
````

Now that we've defined both our variables, our code works as intended.

## Common types of errors and what you did to cause them:


*   NameError - you tried referencing a variable that isn't defined yet
*   IndexError - you tried accessing an index that is out of range for a list or other object (what a flop...)
*   SyntaxError - you're prob missing a parentheses
*   TypeError - you tried using a function on an object of the wrong type (e.g. trying to compute ``len(6)``)
*   ValueError - you tried using a function on an object with a weird value (e.g. ``int("dog")``)
*   AttributeError - u probably didnt pass self to your class method or are trying to access an attribute or method that you didn't define

For more info on the specific error types, the [Error Encyclopedia](https://cs.carleton.edu/cs_comps/1213/pylearn/final_results/encyclopedia/) is a really good resource.

# The ULTIMATE Debugging Guide

## WTF are bugs?

Bugs are a catch-all term for problems in code. Sometimes they refer to errors (e.g. IndexError), or sometimes your code doesn't break but it doesn't do what you wanted it to do. In both scenarios, we'd refer to that code as "buggy."

````python
def complicated_function(my_list, my_other_list):
    '''
    Return a merged version of the two input lists by grouping together elements that share indices
    '''

    out_list = []
    for idx in range(len(my_list)):
        out_list.append(my_list[idx])
        out_list.append(my_other_list[idx])

    return out_list

complicated_function([1, 3, 5], [2, 4, 6])
````

This code is buggy, but it's hard to tell why. When we passed in those two lists, the code worked exactly as intended. But watch what happens when we give the function a different input.

````python
complicated_function([1, 3, 5], [2, 4, 6, 8])
````

The code worked, but did it work as intended? The output is clearly missing the element `8`, so something's awry in our `complicated_function()`.

````python
complicated_function([1, 3, 5], [2, 4])
````

And in this other test case, our function doesn't even run properly! By trying out different test cases, we can see if our code is working as intended or if there are bugs.

Now, we're going to walk you through the ultimate foolproof debugging process.

## Step 1: Add prints

````python
def complicated_function(my_list, my_other_list):
    '''
    Return a merged version of the two input lists by grouping together elements that share indices
    '''

    out_list = []
    for idx in range(len(my_list)):

        print(f'idx = {idx}, len of out_list = {len(out_list)}')

        out_list.append(my_list[idx])
        out_list.append(my_other_list[idx])

    print(f'final len of out_list = {len(out_list)}')
    return out_list

complicated_function([1, 3, 5], [2, 4, 6])
````

The first step to any debugging process is to **GET INFORMATION**. We do this in Python by adding print statements to essentially monitor what is happening in our for loop. Generally, any time you have a bug involving a loop or conditional statement, printing out the value of the relevant variables is never harmful. Just be sure to format using f-strings so that you can understand what variables you're monitoring and the value they take.

Clearly in this first example case, everything ran as intended. We expect the output to have 6 elements since we're merging two lists with 3 elements. Let's see what happens on the two buggy test cases.

## Step 2: Think like a detective

````python
complicated_function([1, 3, 5], [2, 4, 6, 8])
````

As we can see, the for loop stops at idx=2, which might raise some suspicion since the 2nd list has 3 indices... perhaps we're accidentally skipping over the last element of the 2nd list somehow.

````python
complicated_function([1, 3, 5], [2, 4])
````

This output is also strange, since in this case we get to idx=2 even though the 2nd list only has elements up to idx=1. Clearly, something's wrong with our idx variable.

Just by adding that one print statement, we can already tell that our issue relates to the way we're indexing elements. 

If we look back at the relevant for loop, we see that we're going from idx=0 to idx=len(my_list) which ignores the size of the second list. Clearly, that isn't going to work, because if the lists have mismatching sizes, either we'll miss the later elements of the 2nd list or try to index elements of the 2nd list when they don't exist.

## Step 3: Make the necessary changes

In the case of merging two lists, we have to think of a better way to go through each element of both lists.

In our case, it's necessary to break down the problem into a few separate steps:
1. Iterate through the MINIMUM length list and add elements like we do before. This way, we'll never run into an IndexError, since the for loop will end when we've exauhsted all the elements in the smaller list
2. If we've gotten to this point, one list has had all of its elements merged, and the other must have some remaining elements (if it doesn't, we've finished our task). Because of this, we can just iterate through every remaining element and add it at the end.

````python
def complicated_function(my_list, my_other_list):
    '''
    Return a merged version of the two input lists by grouping together elements that share indices
    '''

    out_list = []

    min_length = min(len(my_list), len(my_other_list))

    for idx in range(min_length):

    # print(f'idx = {idx}, len of out_list = {len(out_list)}')

        out_list.append(my_list[idx])
        out_list.append(my_other_list[idx])

    if len(my_list) == min_length: # this means that the remaining elements are in my_other_list
        for idx in range(min_length, len(my_other_list)):
            out_list.append(my_other_list[idx])

    else: # this means that the remaining elements are in my_list
        for idx in range(min_length, len(my_list)):
            out_list.append(my_list[idx])

    return out_list

complicated_function([1, 3, 5], [2, 4, 6])
````

````python
complicated_function([1, 3, 5], [2, 4])
````

````python
complicated_function([1, 3, 5], [2, 4, 6, 8])
````

# Best Coding Practices

## Variable Naming

### Imagine you're the owner of a cookie store, and you're trying to save customer information as a set of a few Python variables. Which variable naming scheme is better?

1. 
```
x = 10
y = 'John Doe'
z = 'loved our valentines day bake sale cookies'
```

2. 
```
age = 10
name = 'John Doe'
notes = 'loved our valentines day bake sale cookies'
```


Spoiler alert: #2 is far more informative. In general, your variable names should **TELL** the user something they don't already know about what the variable represents. 

This doesn't mean to be overly specific: if you're going to be typing out the name of a variable a lot, you want to avoid making the name too long. But also don't leave it as something uninformative like `xx` or `var1`.


---


## Avoid redundancy

### What's wrong with this variable name?
`name_str = 'John Doe'` 

...

...

...

This variable name sucks because we already know `'John Doe'` is a string. The `str` part of this variable name is redundant. As such, we want to avoid these types of redundant names.

---


## Avoid mental maps

### What's wrong with this variable name?
`state_dict = {'California': 10, 'Nebraska': 4}`

...

...

...

Not only is this variable name redundant, but it also doesn't tell us about the information it stores. It's easy to tell that the keys are states, but what are the values? This variable name doesn't tell us that. A far better name would be something like `users_by_state` if we're counting users across the US.

## Comments and Docstrings

````python
def concatenate(list1, list2):
    '''
    Concatenate two lists into one, with list1 occupying the left elements and list2 occupying the right elements.

    Input:
    - list1 (List): first list of elements you want to merge, will be on the left of concatenated list
    - list2 (List): second list of elements you want to merge, will be on the right of concatenated list

    Returns:
    - concat (List): concatenation of list1 (left) and list2 (right)
    '''

    concat = list1
    concat.extend(list2)

    return concat

testcase_left = [1, 2, 3]
testcase_right = [4, 5, 6]

concatenate(testcase_left, testcase_right)
````

In the above example, this function concatenates two lists together into one. The docstring (text surrounded by quote marks) helps inform the user what the function does, what it takes as input, and what the outputs look like. Notice how the type of each parameter is specified, so if a bug occurs, the user can double check that the type of each parameter matches what's expected in the docstring. Python also has the capability to enforce typing, but we won't cover that in this lesson.

## Rule of thumb for comments: if your code is unreadable without them, you probably need to make your variable and function names clearer.

This isn't true in every case, but often times comments serve to explicitly highlight or justify the use of certain functions, methods, or logical expressions. Comments should just be there to help you understand the code you wrote. But if they're the only thing making your code "readable," perhaps you need to re-evaluate how you're naming everything.

```
if len(my_list) == min_length: # this means that the remaining elements are in my_other_list
    for idx in range(min_length, len(my_other_list)):
        out_list.append(my_other_list[idx])

else: # this means that the remaining elements are in my_list
    for idx in range(min_length, len(my_list)):
        out_list.append(my_list[idx])
```

In the `complicated_function()` example from earlier, these comments worked well because they explicitly highlighted the logic in this if/else block. You wouldn't be able to tell the assumptions being made just by reading the `len(my_list) == min_length` term, so the comments are there to clarify that logic. Note that this code isn't super clean, and there are ways it can be made clearer, but this example is primarily to highlight what good comments look like.

## Why did this code segment work? We didn't define `x` explicitly in this code block.

Notebooks work with something called a `kernel` which stores all the variable and session data while you're using it. If you were to restart this notebook and run the code cell again, it would throw a `NameError`.

Often times, a common source of error is naming a variable and having it saved to your kernel, and changing its name in the variable declaration but *NOT* anywhere that it's referenced. As a final tip, whenever you suspect you're having issues with your kernel, try restarting it and running all cells just to be sure your bugs aren't a problem with your notebook.
